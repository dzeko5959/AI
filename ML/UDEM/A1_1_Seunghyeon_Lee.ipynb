4. Select one gender to visualize the relationship between dependent and other variables.

```python
import matplotlib.pyplot as plt

df_Male = df[df["Sexo"] == "Male"]

plt.figure(figsize=(8, 5))
plt.scatter(df_Male["ComeVegetales"], df_Male["NivelDeObesidad"].astype('category').cat.codes, alpha=0.1, color='skyblue')
plt.xlabel("ComeVegetales (average)")
plt.ylabel("Nivel de Obesidad (Categorical)")
plt.title("Relaci√≥n entre el consumo de vegetales y el nivel de obesidad (hombres)")
plt.show()
```

```python
import matplotlib.pyplot as plt

df_female = df[df["Sexo"] == "Female"]

plt.figure(figsize=(8, 5))
plt.scatter(df_female["ComeVegetales"], df_female["NivelDeObesidad"].astype('category').cat.codes, alpha=0.1, color='pink')
plt.xlabel("ComeVegetales (average)")
plt.ylabel("Nivel de Obesidad (Categorical)")
plt.title("Relaci√≥n entre el consumo de vegetales y el nivel de obesidad (mujeres)")
plt.show()
```
